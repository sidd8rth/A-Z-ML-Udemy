```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Position_Salaries.csv")
X = dataset.iloc[:,1:-1].values
Y = dataset.iloc[:,-1].values
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression()
linear_regressor.fit(X,Y)
```

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly_mat = PolynomialFeatures(degree=4)
X_poly = poly_mat.fit_transform(X)
print(X_poly)
polynomial_regressor = LinearRegression()
polynomial_regressor.fit(X_poly,Y)
```

# Plotting Linear Regression

```python
plt.scatter(X,Y,color="red")
plt.plot(X,linear_regressor.predict(X),"b")
plt.show()
```

```python
plt.scatter(X,Y,color="red")
plt.plot(X,polynomial_regressor.predict(X_poly),"b")
plt.show()
```

```python
plt.scatter(X,Y,color="red")
plt.plot(X,polynomial_regressor.predict(X_poly),"b")
plt.show()
```

```python
plt.scatter(X,Y,color="red")
plt.plot(X,polynomial_regressor.predict(X_poly),"b")
plt.show()
```

```python
X_grid = np.arange(np.min(X),np.max(X),0.1)
X_grid = X_grid.reshape(len(X_grid),1)
plt.scatter(X,Y,color="red")
plt.plot(X_grid,polynomial_regressor.predict(poly_mat.fit_transform(X_grid)),"b")
plt.show()
```
