```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Mall_Customers.csv")
X = data.iloc[:,[3,4]].values
```

```python
#print(X)
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X,method='ward'))
plt.title("Dendrogram")
plt.xlabel("Customers")
plt.ylabel("Euclidean Distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
y_pred = hc.fit_predict(X)
```

```python
print(y_pred)
```

```python
plt.scatter(X[y_pred == 0, 0], X[y_pred==0,1],s=50, c='red', label='Cluster 1')
plt.scatter(X[y_pred == 1, 0], X[y_pred==1,1],s=50, c='blue', label='Cluster 2')
plt.scatter(X[y_pred == 2, 0], X[y_pred==2,1],s=50, c='black', label='Cluster 3')
plt.scatter(X[y_pred == 3, 0], X[y_pred==3,1],s=50, c='yellow', label='Cluster 4')
plt.scatter(X[y_pred == 4, 0], X[y_pred==4,1],s=50, c='green', label='Cluster 5')
#plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='magenta', label='Centroids')
plt.title("Clusters")
plt.xlabel("Income")
plt.ylabel('Spending')
plt.legend()
plt.show()
```
