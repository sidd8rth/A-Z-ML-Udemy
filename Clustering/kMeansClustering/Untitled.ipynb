# Importing Lib.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Data Loading

```python
data = pd.read_csv("Mall_Customers.csv")
X = data.iloc[:,[3,4]].values
```

```python
# print(X)
```

# Elbow Method

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss)
#plt.xticks(i for i in range())
plt.show()
```

```python
# From here we can see cluster = 5
```

# Training

```python
kmeans = KMeans(n_clusters=5,init='k-means++', random_state=42)
y_pred = kmeans.fit_predict(X)
```

```python
print(y_pred)
print(kmeans.cluster_centers_)
```

# Plotting

```python
plt.scatter(X[y_pred == 0, 0], X[y_pred==0,1],s=50, c='red', label='Cluster 1')
plt.scatter(X[y_pred == 1, 0], X[y_pred==1,1],s=50, c='blue', label='Cluster 2')
plt.scatter(X[y_pred == 2, 0], X[y_pred==2,1],s=50, c='black', label='Cluster 3')
plt.scatter(X[y_pred == 3, 0], X[y_pred==3,1],s=50, c='yellow', label='Cluster 4')
plt.scatter(X[y_pred == 4, 0], X[y_pred==4,1],s=50, c='green', label='Cluster 5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='magenta', label='Centroids')
plt.title("Clusters")
plt.xlabel("Income")
plt.ylabel('Spending')
plt.legend()
plt.show()
```
